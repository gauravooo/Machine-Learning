Multi-Feature Dataset

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([
    [1,1,0],
    [2,1,1],
    [3,2,1],
    [5,2,2],
    [8,3,3]
])

y = np.array([45000, 52000, 68000, 85000, 115000])

model = LinearRegression()
model.fit(X, y)

print(f"Intercept (Base Salary): {model.intercept_:.2f}")
print(f"Coefficients (Weights): {model.coef_}")

new_candidate = np.array([[4,2,2]])
prediction = model.predict(new_candidate)
print(f"\nPrediction for new candidate: {prediction[0]:.2f}")
```

Concept of underfitting and overfitting

Underfitting happens when your model is too simple to capture the underlying pattern of the data.
