```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Fake actuals and predictions
y_true = np.array([50000, 60000, 75000, 90000])
y_pred = np.array([52000, 58000, 76000, 85000]) # The model's guesses

# Calculate Metrics
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred)) # Notice the np.sqrt()
r2 = r2_score(y_true, y_pred)

print(f"MAE: ${mae:,.2f}")
print(f"RMSE: ${rmse:,.2f}")
print(f"R-squared: {r2:.4f}")
```

R-squared close to 1 means our model is doing good.
